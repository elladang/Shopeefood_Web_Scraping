```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from my_libs import *
import pandas as pd
import os 
import requests
import time 
```

```python
driver = webdriver.Chrome()
driver.maximize_window()
url = "https://shopeefood.vn/ho-chi-minh/danh-sach-dia-diem-giao-tan-noi"
driver.get(url)
```

```python
list_link_cua_hang= []
load_more = driver.find_elements(by=By.XPATH,value = "//ul[@class='pagination']//a")
for i in load_more[1:-2]: 
    i.click()
    tag_cua_hang = driver.find_elements(by=By.XPATH, value="//div[@class='list-restaurant']//a[@class='item-content']")
    for link_cua_hang in tag_cua_hang:
        link_store = link_cua_hang.get_attribute('href')
        list_link_cua_hang.append(link_store)
    print(list_link_cua_hang)
    print(len(list_link_cua_hang))
    time.sleep(20)
link_shopee_txt= "links_shopee.txt"
write_txt(link_shopee_txt,list_link_cua_hang)
```

```python
list_link_cua_hang_test= [ 'https://shopeefood.vn/ho-chi-minh/lau-thai-pasi','https://shopeefood.vn/ho-chi-minh/an-vat-em-guot','https://shopeefood.vn/ho-chi-minh/co-quan-com-mi-tron-xa-xiu-tra-trai-cay']
list_restaurant_name=[]
list_restaurant_address = []
list_operation_time =[]
list_price_range = []
count = 0

driver_detail = webdriver.Chrome()
driver_detail.maximize_window()
for link in list_link_cua_hang: 
    url = link
    print(url)
    driver_detail.get(url)
    try:
        ok_button = driver_detail.find_element(by=By.XPATH, value='//button[@class= "btn btn-red align-right"]')
        ok_button.click()
        time.sleep(10)
    except:
        pass
        time.sleep(30)
        
    try: 
        # restaurant_name
        restaurant_name_tag = driver_detail.find_element(by=By.XPATH, value="//h1[@class='name-restaurant']")
        restaurant_name = restaurant_name_tag.text
        print(restaurant_name)
        list_restaurant_name.append(restaurant_name)

        # adress

        restaunrant_address = driver_detail.find_element(by=By.XPATH, value="//div[@class='address-restaurant']")
        restaunt_address = restaunrant_address.text
        print(restaunt_address)
        list_restaurant_address.append(restaunt_address)

        # operation_time
        try: 
            operation_time_tag = driver_detail.find_element(by=By.XPATH, value="//div[@class='time']")
            op_time = operation_time_tag.text
            print(op_time)
            list_operation_time.append(op_time)
        except: 
            op_time = "Đã đóng"
            print(op_time)
            list_operation_time.append(op_time)

        # price_range

        price_range_tag =  driver_detail.find_element(by=By.XPATH, value="//div[@class='cost-restaurant']")
        pr = price_range_tag.text
        print(pr)
        list_price_range.append(pr)
        #  Dem 
        count +=1
        print(f"Restaurant{count} : {restaurant_name}")
        time.sleep(30)
    except Exception as e:
        pass


# Export to CSV
df_products = pd.DataFrame({
    "id":[i for i in range(1,len(list_restaurant_name)+1)],
    "Restaurant": list_restaurant_name,
    "Adress": list_restaurant_address,
    "Time": list_operation_time,
    "Price": list_price_range,
    "City": "HCM"
})
df_products.to_csv("restaurant_shopeefood.csv", index=False)
# Dong Driver
driver_detail.close()
```
